```python
from PIL import Image

import json
import cv2 
import os
import glob 
from os import path

#reading from json file
img_dir = "C:/Users/monsur/Desktop/LearningPytorch/converted"
data_path = os.path.join(img_dir,'*g')
files = glob.glob(data_path) 
#print(files)
file=[]
img_lbl=[]
for i in range(len(files)):
    file.append(files[i])
    #print(file[0])
    inp_file=str(file[0])
    #print(inp_file)
    inp_file = inp_file.lstrip('C:/Users/monsur/Desktop/LearningPytorch/converted\ ')
    inp_file = inp_file.rstrip('.jpg')
    #print(inp_file)
    img_lbl.append(inp_file)
    #print(img_lbl)
    file=[]
    
print(img_lbl)    
```

```python

#extracting labels and points
for idx in range(len(img_lbl)):
    mkdirf= 'C:/Users/monsur/Desktop/croppingimage/newcrp/'+str(img_lbl[idx])
    os.makedirs(mkdirf)
    point=[]
    labels=[]
    json_name = img_lbl[idx]+'.json'
    #print(json_name)
    json_opndir = path.join('C:/Users/monsur/Desktop/LearningPytorch/converted/',json_name)
    print(json_opndir)
    with open(json_opndir,encoding="utf8") as json_file:
        data = json.load(json_file)
        for p in data['shapes']:
            point.append(p['points'])
            labels.append(p['label'])

    img_name = img_lbl[idx]+'.jpg'
    img_opndir = path.join('C:/Users/monsur/Desktop/LearningPytorch/converted/',img_name)
    img = Image.open(img_opndir)
    print(idx)
    print(img_opndir)
    arr=[]

    for index in range(len(labels)):
        for i in range(len(point[index])):
            for j in range(len (point[0][0])):
                arr.append((point[index][i][j]))
                #print(arr)
        area = (arr[0],arr[1],arr[2],arr[3])
        #print("area inserted successfully")
        arr=[]
        cropped_img = img.crop(area)
            
    
        crop_rename =str(labels[index])+'.jpg'
        crop_rename = crop_rename.replace('*','')
        crop_rename = crop_rename.replace('\"','')
        crop_rename = crop_rename.replace('?','')
        crop_rename = crop_rename.replace('/','')
        crop_rename = crop_rename.replace('|','')
        print(crop_rename)
        
        #print(crop_rename)
        
        imgfolder=img_lbl[idx] 
        
        crop_rename = imgfolder+'/'+crop_rename
        img_dir = path.join('C:/Users/monsur/Desktop/croppingimage/newcrp/',crop_rename)
        #if(img_dir == 'C:/Users/monsur/Desktop/croppingimage/newcrp/*.jpg'):
           #i=1 
        #else:
        try:
            cropped_img = cropped_img.save(img_dir)
        except:
            print("Error")
            
    
    
    
    
    
```
